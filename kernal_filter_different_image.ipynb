**KERNAL FILTER MANIPULATION**<br>
kernal/filter/sliding window are ~same

```python
image = cv2.imread('image.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(image,cmap="gray")
```

```python
#FILTER CHANGE HI INTENSITY HOTSPOTS TO LO INTENSITY VALUES
import cv2
import numpy as np
import matplotlib.pyplot as plt

def apply_kernel(image):
    output_image = image.copy()
    height, width = image.shape
    for i in range(1, height-1):
        for j in range(1, width-1):
            kernel = image[i-1:i+2, j-1:j+2]
            avg_value = np.mean(kernel)
            if avg_value >= 200:
                output_image[i, j] = 0

    return output_image

image = cv2.imread('image.jpg', cv2.IMREAD_GRAYSCALE)
output_image = apply_kernel(image)
plt.imshow(output_image,cmap="gray")
```

```python
#sharping kernal
import numpy as np
import matplotlib.pyplot as plt
import cv2

image = cv2.imread("image.jpg", cv2.IMREAD_GRAYSCALE)

kernel = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
image = cv2.filter2D(image, -1, kernel)

plt.imshow(image,cmap="gray")
plt.show()
```

```python
#mean filter
import cv2
import numpy as np

image = cv2.imread('image.jpg', cv2.IMREAD_GRAYSCALE)

kernel_size = 3
kernel = np.ones((kernel_size, kernel_size), np.float32) / (kernel_size * kernel_size)

filtered_image = cv2.filter2D(image, -1, kernel)

# Display the original and filtered image
plt.imshow(filtered_image,cmap="gray")
plt.show()
```

```python
#Prewitt Operator [X-axis]
import numpy as np
import matplotlib.pyplot as plt
import cv2

image = cv2.imread("image.jpg", cv2.IMREAD_GRAYSCALE)

kernel = np.array([[-1,0,1], [-1,0,1], [-1,0,1]])
image = cv2.filter2D(image, -1, kernel)

plt.imshow(image,cmap="gray")
plt.show()
```
