```python
import numpy as np

image =cv2.imread("image.jpg",cv2.IMREAD_GRAYSCALE)
matA=np.array(image)
matA=matA.T
matA=matA[:,::-1]
plt.imshow(matA,cmap="gray")
plt.show()
```
